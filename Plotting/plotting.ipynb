```python
# 📊 Comparing PPO Training Metrics Between Runs

import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# For Jupyter notebooks, get the directory containing the notebook
notebook_dir = os.getcwd()

# Paths to the metrics folders
run1_dir = os.path.join(notebook_dir, "run_default_1m", "data")
run2_dir = os.path.join(notebook_dir, "run_default_500k", "data")

# Create a folder for saving comparison figures
comparison_dir = os.path.join(notebook_dir, "comparison_figures")
os.makedirs(comparison_dir, exist_ok=True)

# Print paths for debugging
print(f"Current working directory: {notebook_dir}")
print(f"Run 1 metrics: {run1_dir}")
print(f"Run 2 metrics: {run2_dir}")
print(f"Saving comparisons to: {comparison_dir}")

# Check if directories exist
for dir_path in [run1_dir, run2_dir]:
    if not os.path.exists(dir_path):
        print(f"⚠️ Directory not found: {dir_path}")

# List of metrics and friendly titles
metric_files = {
    'ep_rew_mean.csv': 'Episode Reward Mean',
    'policy_gradient_loss.csv': 'Policy Gradient Loss',
    'approx_kl.csv': 'Approximate KL Divergence',
    'entropy_loss.csv': 'Entropy Loss',
    'explained_variance.csv': 'Explained Variance'
}

# Compare each metric
for filename, title in metric_files.items():
    run1_path = os.path.join(run1_dir, filename)
    run2_path = os.path.join(run2_dir, filename)
    
    # Check if both files exist
    if not os.path.exists(run1_path):
        print(f"⚠️ File not found: {run1_path}")
        continue
    if not os.path.exists(run2_path):
        print(f"⚠️ File not found: {run2_path}")
        continue
    
    # Load data
    df1 = pd.read_csv(run1_path)
    df2 = pd.read_csv(run2_path)
    
    # Check if data has expected columns
    if not ('Step' in df1.columns and 'Value' in df1.columns):
        print(f"⚠️ File {run1_path} does not have expected columns ['Step', 'Value']")
        continue
    if not ('Step' in df2.columns and 'Value' in df2.columns):
        print(f"⚠️ File {run2_path} does not have expected columns ['Step', 'Value']")
        continue
    
    # Calculate statistics
    mean1 = df1['Value'].mean()
    mean2 = df2['Value'].mean()
    std1 = df1['Value'].std()
    std2 = df2['Value'].std()
    max1 = df1['Value'].max()
    max2 = df2['Value'].max()
    min1 = df1['Value'].min()
    min2 = df2['Value'].min()
    
    # Create comparison figure
    plt.figure(figsize=(12, 6))
    # Use a better color palette
    color1 = '#1f77b4'  # A nice blue
    color2 = '#ff7f0e'  # A contrasting orange


    # Plot both metrics on the same plot
    plt.plot(df1['Step'], df1['Value'], label='1M Steps', color=color1)
    plt.plot(df2['Step'], df2['Value'], label='500K Steps', color=color2)
    
    plt.title(f'Comparison of {title}')
    plt.xlabel('Step')
    plt.ylabel(title)
    plt.grid(True)
    plt.legend()
    
    # # Add statistical information as text
    # stats_text = (
    #     f"1M Steps: Mean={mean1:.4f}, Std={std1:.4f}, Min={min1:.4f}, Max={max1:.4f}\n"
    #     f"500K Steps: Mean={mean2:.4f}, Std={std2:.4f}, Min={min2:.4f}, Max={max2:.4f}"
    # )
    # plt.figtext(0.5, 0.01, stats_text, ha='center', fontsize=10, 
    #             bbox={'facecolor': 'white', 'alpha': 0.8, 'pad': 5})
    
    plt.tight_layout()
    
    # Save comparison figure
    fig_path = os.path.join(comparison_dir, f"comparison_{filename.replace('.csv', '')}.png")
    plt.savefig(fig_path, dpi=300)
    print(f"✅ Saved comparison to: {fig_path}")
    
    # Display in notebook
    plt.show()

    # Print statistical comparison
    print(f"\n📊 Statistical comparison for {title}:")
    print(f"  1M Steps: Mean={mean1:.4f}, Std={std1:.4f}, Min={min1:.4f}, Max={max1:.4f}")
    print(f"  500K Steps: Mean={mean2:.4f}, Std={std2:.4f}, Min={min2:.4f}, Max={max2:.4f}")
    print(f"  Difference (1M - 500K): Mean={mean1-mean2:.4f}, Max={max1-max2:.4f}")
    print("-" * 50)

print("\n✅ All comparisons have been completed and saved to:", comparison_dir)
```

```python
# 📊 Comparing Default vs Tuned PPO Training Metrics

import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# For Jupyter notebooks, get the directory containing the notebook
notebook_dir = os.getcwd()

# Paths to the metrics folders
run1_dir = os.path.join(notebook_dir, "run_default_1m", "data")
run2_dir = os.path.join(notebook_dir, "run_tuned_1m", "data")

# Create a folder for saving comparison figures
comparison_dir = os.path.join(notebook_dir, "default_vs_tuned_comparison")
os.makedirs(comparison_dir, exist_ok=True)

# Print paths for debugging
print(f"Current working directory: {notebook_dir}")
print(f"Default model metrics: {run1_dir}")
print(f"Tuned model metrics: {run2_dir}")
print(f"Saving comparisons to: {comparison_dir}")

# Check if directories exist
for dir_path in [run1_dir, run2_dir]:
    if not os.path.exists(dir_path):
        print(f"⚠️ Directory not found: {dir_path}")

# List of metrics and friendly titles
metric_files = {
    'ep_rew_mean.csv': 'Episode Reward Mean',
    'policy_gradient_loss.csv': 'Policy Gradient Loss',
    'approx_kl.csv': 'Approximate KL Divergence',
    'entropy_loss.csv': 'Entropy Loss',
    'explained_variance.csv': 'Explained Variance'
}

# Compare each metric
for filename, title in metric_files.items():
    run1_path = os.path.join(run1_dir, filename)
    run2_path = os.path.join(run2_dir, filename)
    
    # Check if both files exist
    if not os.path.exists(run1_path):
        print(f"⚠️ File not found: {run1_path}")
        continue
    if not os.path.exists(run2_path):
        print(f"⚠️ File not found: {run2_path}")
        continue
    
    # Load data
    df1 = pd.read_csv(run1_path)  # Default model
    df2 = pd.read_csv(run2_path)  # Tuned model
    
    # Check if data has expected columns
    if not ('Step' in df1.columns and 'Value' in df1.columns):
        print(f"⚠️ File {run1_path} does not have expected columns ['Step', 'Value']")
        continue
    if not ('Step' in df2.columns and 'Value' in df2.columns):
        print(f"⚠️ File {run2_path} does not have expected columns ['Step', 'Value']")
        continue
    
    # Calculate statistics
    mean1 = df1['Value'].mean()
    mean2 = df2['Value'].mean()
    std1 = df1['Value'].std()
    std2 = df2['Value'].std()
    max1 = df1['Value'].max()
    max2 = df2['Value'].max()
    min1 = df1['Value'].min()
    min2 = df2['Value'].min()
    
    # Create comparison figure
    plt.figure(figsize=(12, 6))
    
    # Use a better color palette
    color1 = '#1f77b4'  # Blue for default
    color2 = '#2ca02c'  # Green for tuned
    
    # Plot both metrics on the same plot
    plt.plot(df1['Step'], df1['Value'], label='Default Model', color=color1)
    plt.plot(df2['Step'], df2['Value'], label='Tuned Model', color=color2)
    
    plt.title(f'Comparison of {title}: Default vs Tuned Model')
    plt.xlabel('Step')
    plt.ylabel(title)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    
    # Enhance the plot with a light background grid
    plt.gca().set_facecolor('#f8f9fa')
    
    # Add improvement percentage if it's episode reward
    if 'reward' in filename.lower():
        improvement = ((max2 - max1) / max1) * 100 if max1 != 0 else 0
        plt.annotate(f'Max Reward Improvement: {improvement:.2f}%', 
                    xy=(0.02, 0.95), xycoords='axes fraction',
                    bbox=dict(boxstyle="round,pad=0.3", fc='white', ec="gray", alpha=0.8))
    
    plt.tight_layout()
    
    # Save comparison figure
    fig_path = os.path.join(comparison_dir, f"comparison_{filename.replace('.csv', '')}.png")
    plt.savefig(fig_path, dpi=300)
    print(f"✅ Saved comparison to: {fig_path}")
    
    # Display in notebook
    plt.show()

    # Print statistical comparison
    print(f"\n📊 Statistical comparison for {title}:")
    print(f"  Default Model: Mean={mean1:.4f}, Std={std1:.4f}, Min={min1:.4f}, Max={max1:.4f}")
    print(f"  Tuned Model: Mean={mean2:.4f}, Std={std2:.4f}, Min={min2:.4f}, Max={max2:.4f}")
    
    # Calculate percentage improvements
    mean_improvement = ((mean2 - mean1) / abs(mean1)) * 100 if mean1 != 0 else 0
    max_improvement = ((max2 - max1) / abs(max1)) * 100 if max1 != 0 else 0
    
    # Different wording based on whether higher or lower values are better
    if 'reward' in filename.lower() or 'explained_variance' in filename.lower():
        mean_sign = "+" if mean_improvement > 0 else ""
        max_sign = "+" if max_improvement > 0 else ""
    else:  # For losses, lower is better
        mean_sign = "+" if mean_improvement < 0 else ""
        max_improvement = -max_improvement  # Invert for losses
        max_sign = "+" if max_improvement > 0 else ""
    
    print(f"  Improvement: Mean={mean_sign}{mean_improvement:.2f}%, Max={max_sign}{max_improvement:.2f}%")
    print("-" * 50)

# Create a summary table of all metrics
print("\n📋 Summary Table of Improvements")
summary_data = []

for filename, title in metric_files.items():
    run1_path = os.path.join(run1_dir, filename)
    run2_path = os.path.join(run2_dir, filename)
    
    if os.path.exists(run1_path) and os.path.exists(run2_path):
        df1 = pd.read_csv(run1_path)
        df2 = pd.read_csv(run2_path)
        
        if 'Value' in df1.columns and 'Value' in df2.columns:
            mean1 = df1['Value'].mean()
            mean2 = df2['Value'].mean()
            max1 = df1['Value'].max()
            max2 = df2['Value'].max()
            
            mean_improvement = ((mean2 - mean1) / abs(mean1)) * 100 if mean1 != 0 else 0
            max_improvement = ((max2 - max1) / abs(max1)) * 100 if max1 != 0 else 0
            
            # Adjust sign for metrics where lower is better
            if not ('reward' in filename.lower() or 'explained_variance' in filename.lower()):
                mean_improvement = -mean_improvement
                max_improvement = -max_improvement
            
            summary_data.append({
                'Metric': title,
                'Default Mean': mean1,
                'Tuned Mean': mean2,
                'Mean Improvement (%)': mean_improvement,
                'Default Max': max1,
                'Tuned Max': max2,
                'Max Improvement (%)': max_improvement
            })

# Create a summary DataFrame
summary_df = pd.DataFrame(summary_data)

# Print the summary table
print(summary_df.to_string(index=False, float_format=lambda x: f"{x:.4f}"))

# Save the summary table
summary_path = os.path.join(comparison_dir, "metrics_improvement_summary.csv")
summary_df.to_csv(summary_path, index=False)
print(f"\n✅ Saved metrics improvement summary to: {summary_path}")

print("\n✅ All comparisons have been completed and saved to:", comparison_dir)
```
