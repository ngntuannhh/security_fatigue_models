```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Define CSV file paths (modify if needed)
file_path = r"C:\Users\Tuan Anh HSLU\OneDrive - Hochschule Luzern\Desktop\HSLU22\Bachelor Thesis\ML Models\Plotting\run-PPO_14-tag-eval_mean_reward.csv"

# Load the CSV data
df = pd.read_csv(file_path)

# Set seaborn style for better visualization
sns.set_theme(style="whitegrid")

# Plot Mean Reward over Timesteps
plt.figure(figsize=(10, 5))
sns.lineplot(x=df["Step"], y=df["Value"], linestyle="-")

# Title and labels
plt.title("Mean Reward Over Training Timesteps")
plt.xlabel("Timesteps")
plt.ylabel("Mean Reward")

# Show the plot
plt.show()
```
